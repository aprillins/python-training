```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Set index and data
index = ['Revenue', 'Distribution expense', 'HR expense', 'Tax', 'Receivable']
mydata = {'amount':[1200000, -200000, -300000, -1200000*0.1, 250000]}

df = pd.DataFrame(data=mydata, index=index)
dfcumsum = df.amount.cumsum()
dfamounttotal = df.sum().amount
df.loc['net'] = dfamounttotal
blank = dfcumsum.shift(1).fillna(0)
blank.loc['net'] = dfamounttotal
max = df.max()

step = blank.reset_index(drop=True).repeat(3).shift(-1)
step[1::3]= np.nan
blank.loc['net'] = 0

myplot = df.plot(kind='bar', stacked=False, bottom=blank, title='Sales', edgecolor='k', linewidth=1) #Plot the bar

neg_offset = max/15
pos_offset = max/30
y_height = df.amount.cumsum().shift(1).fillna(0)

#Start loop for annotation on each bar
loop = 0
for index, row in df.iterrows():
    if row['amount'] == dfamounttotal:
        y = y_height[loop]
    else:
        y = y_height[loop] + row['amount']

    if row['amount'] > 0:
        y += pos_offset
    else:
        y -= neg_offset

    myplot.annotate("{:,.0f}".format(row['amount']), (loop,y), ha="center")
    loop += 1
#End loop

plot_offset = int(max / 5)
myplot.set_ylim(0,blank.max()+int(plot_offset)) #Set y-axis value top limit to give space for the annotation
myplot.plot(step.index, step.values, 'k--', linewidth=1)
```
