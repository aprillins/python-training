```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
fig.suptitle('No axes in this figure')
fig, ax_1st = plt.subplots(2, 2)
plt.subplot(223)
plt.xlabel('borooooo')
plt.title('sup')
```

```python
x = np.linspace(0, 2, 50)
y = np.array([-2])
print(x)
plt.suptitle('No one can rule the world')
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.plot(y, y, label='test')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

plt.legend()

plt.show()
```

```python
x = np.arange(0, 10, 0.2)
y = np.sin(x*0.5)
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
z = np.linspace(1, 0.3, 3)
y *= z
print(z, y)
plt.plot(z, 'go', y, 'r--')
plt.axis([-3,3,-1,2])
plt.xlabel('none')
plt.ylabel('zone')
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(10, 13))
plt.subplot(2,2,1)
plt.plot(names, values)
plt.title('circle')
plt.title('triangle')
plt.subplot(2,2,3)
plt.plot(names,values, 'b^')
plt.subplot(2,2,4)
plt.bar(names, values)
plt.title('bar')
plt.suptitle('Test')
plt.subplot(222)
plt.title('dashed')
plt.plot(names,values, 'd--')
plt.xlabel('bro')
plt.show()
```
