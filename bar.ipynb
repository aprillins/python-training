```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FuncFormatter

data = {'Jakarta Barat': 100000, 'Jakarta Utara': 150000, 'Jakarta Selatan': 120000, 'Jakarta Timur': 90000, 'Jakarta Pusat': 180000}
data_values = list(data.values())
data_keys = list(data.keys())
data_mean = np.mean(data_values)

fig, ax = plt.subplots()

plt.style.use('seaborn-notebook')
ax.barh(data_keys,data_values)
labels = ax.get_xticklabels()
plt.setp(labels, rotation='horizontal', horizontalalignment='left')
plt.rcParams.update({'figure.autolayout': True})
#dir(plt)
#plt.title('Daily Sales of Chicken in DKI Jakarta')
#plt.ylabel('Location')
#plt.xlabel('Kilogram')

ax.title.set(y=1.05)
ax.text(np.max(data_values), 0, "New Company", fontsize=10, verticalalignment="center")
ax.set(title='Daily Sales of Chicken in DKI Jakarta', xlabel='Kilogram', ylabel='Location')
ax.axvline(data_mean, ls='--', color='r')
```
